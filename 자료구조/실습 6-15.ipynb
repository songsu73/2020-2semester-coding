```python
from typing import MutableSequence


def _merge_sort(a: MutableSequence, left: int, right: int) -> None:
        if left < right: ## 병합정렬을 실시할 조건 검사
            center = (left + right) // 2

            _merge_sort(a, left, center)            # 배열 앞부분을 병합 정렬
            _merge_sort(a, center + 1, right)       # 배열 뒷부분을 병합 정렬  feat.재귀호출

            p = j = 0
            i = k = left

            while i <= center: ## a의 앞부분을 buff에 그대로 옮겨놓기 위함
                 buff[p] = a[i]
                 p += 1 ##i와 p를 같이 증가시킴
                 i += 1

            while i <= right and j < p:
                 if buff[j] <= a[i]:
                     a[k] = buff[j]  ##buff에서 원소를 꺼내서 a에 병합
                     j += 1 ##buff 포인터 증가
                 else:
                     a[k] = a[i] ## a[i]가 더 작으므로 a[i]를 먼저 a에 넣는다
                     i += 1 ##a 포인터 증가
                 k += 1 ##원소를 넣을때마다 a에 넣을 포인터 증가

            while j < p: ##a[i]에서 더이상 넣을 원소가 없고 buff에서 원소가 남았을경우 작동하는 반복문
                a[k] = buff[j] ##buff의 나머지 원소를 배열 a에 복사
                k += 1
                j += 1

if __name__ == '__main__':
    print('병합 정렬을 수행합니다')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num    # 원소 수가 num인 배열을 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))
     
    n = len(x)
    buff = [None] * n          
    _merge_sort(x, 0, n - 1)   
    del buff    
    
    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```
