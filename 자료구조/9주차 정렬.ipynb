# 실습 6-1 버블정렬알고리즘

```python
from typing import MutableSequence

def bubble_sort(a:MutableSequence) -> None:
    n = len(a)
    for i in range(n-1) :
        for j in range(n-1,i,-1):
            if a[j-1] > a[j] :
                a[j-1],a[j] = a[j],a[j-1]
                
if __name__ == '__main__' :
    print('버블 정렬을 수행합니다')
    num = int(input('원소 수를 입력하세요 : '))
    x = [None] * num
    
    for i in range(num) :
        x[i] = (input(f'x[{i}] :'))
        
    bubble_sort(x)
    
    print('오름차순으로 정렬합니다 ')
    for i in range(num) :
        print(f'x[{i}] = {x[i]}')
```

# 6-4 버블 정렬의 개선판 2

```python
from typing import MutableSequence

def bubble_sort(a: MutableSequence) -> None:
    """버블 정렬(스캔 범위를 제한)"""
    n = len(a)
    k = 0
    scnt = 0
    ccnt = 0
    while k < n - 1:
        last = n - 1
        for j in range(n - 1, k, -1):
            scnt += 1
            if a[j - 1] > a[j]:
                a[j - 1], a[j] = a[j], a[j - 1]
                last = j
                ccnt += 1
        k = last
    print(f'{ccnt}만큼 교환하였습니다')
    print(f'{scnt}만큼 비교하였습니다')

if __name__ == '__main__':
    print('버블 정렬을 합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num    

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    bubble_sort(x)      

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```

# 실습 6-5 셰이커 정렬

```python
from typing import MutableSequence

def shaker_sort(a: MutableSequence) -> None:
    left = 0
    right = len(a) - 1
    last = right
    ccnt = 0
    scnt = 0
    while left < right:
           
        for j in range(right, left, -1):
            ccnt += 1
            if a[j - 1] > a[j]:
                a[j - 1], a[j] = a[j], a[j - 1]
                last = j
                scnt += 1
        left = last
       
        
        for j in range(left, right):
         
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
                last = j
                scnt += 1
        right = last
       
    print(f"비교를 {ccnt}번 했습니다.")
    print(f"교환을 {scnt}번 했습니다.")
if __name__ == '__main__':
    print('셰이커 정렬을 수행합니다')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}] : '))

    shaker_sort(x) 

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
```

# 실습 6-6 단순 선택 정렬 함수

```python
from typing import MutableSequence

def selection_sort(a: MutableSequence) -> None:
    n = len(a)
    ccnt = 0
    scnt = 0
    for i in range(n - 1):
        min = i  
        for j in range(i + 1, n): 
            ccnt += 1
            if a[j] < a[min]:
                min = j
               
        a[i], a[min] = a[min], a[i]
        scnt += 1
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')
if __name__ == '__main__':
    print('단순 선택 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}] : '))

    selection_sort(x)

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        ## 비교횟수는 n(n-1) / 2
```

# 실습 6-7 단순 삽입 정렬

```python
from typing import MutableSequence

def insertion_sort(a: MutableSequence) -> None :
    n = len(a)
    for i in range(1, n):
        j = i
        tmp = a[i]
        while j > 0 and a[j - 1] > tmp:
            a[j] = a[j - 1]
            j -= 1
        a[j] = tmp

if __name__ == '__main__':
    print('단순 삽입 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    insertion_sort(x)  

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        ##비교횟수와 교환횟수는 n^2 / 2 번
        
```

# 6-8 셀 정렬을 수행하는 프로그램

```python
from typing import MutableSequence

def shell_sort(a: MutableSequence) -> None:
    n = len(a)
    h = n // 2
    while h > 0:
        for i in range(h, n):
            j = i - h
            tmp = a[i]
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]
                j -= h
            a[j + h] = tmp
        h //= 2

if __name__ == '__main__':
    print('셸 정렬을 수행합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    shell_sort(x) 

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        ## h의 초기값은 n // 2 while문을 반복할때마다 반복 0이될때까지
```

# 실습 6-9 셸정렬 (배수가 되지 않도록 h = h *3 + 1의 배열 생성)

```python
from typing import MutableSequence

def shell_sort(a: MutableSequence) -> None:
    n = len(a)
    h = 1

    while h < n // 9: ## h*3+1 배열의 크기를 제한하는 역할
        h = h * 3 + 1

    while h > 0:
        for i in range(h, n):
            j = i - h
            tmp = a[i]
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]
                j -= h
            a[j + h] = tmp
        h //= 3

if __name__ == '__main__':
    print('셸 정렬을 수행합니다(h * 3 + 1의 수열 사용).')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num  

    for i in range(num):
        x[i] = int(input(f'x[{i}]: '))

    shell_sort(x)  

    print('오름차순으로 정렬했습니다.')
    for i in range(num):
        print(f'x[{i}] = {x[i]}')
        ## 셸 정렬을 효과적으로 수행하기 위해 h값을 변형한 정렬
```
