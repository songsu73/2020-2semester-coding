#  Matplotlib 라이브러리 사용

```python
tips = sns.load_dataset("tips") 
print(tips.head())
print(type(tips)) ##tips 데이터셋 호출
```

```python
fig = plt.figure() 
axes1 = fig.add_subplot(1, 1, 1)  ##빈 plot 생성
```

```python
axes1.hist(tips['total_bill'], bins=10) 
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency') 
axes1.set_ylabel('Total Bill') 
 ##plot에   title, xlabel, ylabel 추가
##히스토그램을 생성하고 구간을 10개로 나눔
fig
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(tips['total_bill'], tips['tip']) 
axes1.set_title('Scatterplot of Total Bill vs Tip') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') ##산점도 그리기
```

```python
boxplot = plt.figure() 
axes1 = boxplot.add_subplot(1, 1, 1) ##빈 박스그래프

axes1.boxplot( 
    [tips[tips['sex'] == 'Female']['tip'], 
     tips[tips['sex'] == 'Male']['tip']], 
    labels=['Female', 'Male']) ##박스그래프 생성


axes1.set_xlabel('Sex') 
axes1.set_ylabel('Tip') 
axes1.set_title('Boxplot of Tips by Sex') ##제목과 xlabel, ylabel 추가
```

```python
def recode_sex(sex): ##recode_sex함수 생성(여자면 False 아니면 True)
    if sex == 'Female':
        return 0 
    else:
        return 1
tips['sex_color'] = tips['sex'].apply(recode_sex)
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1)  ##빈 산점도 그래프
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 10,    
    c=tips['sex_color'],
    alpha=0.5) ##산점도 그래프 생성

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

# Seaborn 라이브러리 사용

```python
import seaborn as sns

tips = sns.load_dataset("tips")
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])  ##히스토그램 그리기
ax.set_title('Total Bill Histogram with Density Plot') 
```

```python
ax = plt.subplots() 
ax = sns.distplot(tips['total_bill'], kde=False) ##밀집도 그래프를 제거하는 kde = False
ax.set_title('Total Bill Histogram') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Frequency') 
```

```python
ax = plt.subplots()  
ax = sns.distplot(tips['total_bill'], hist=False) ##밀집도 그래프만 나타내는 hist = False
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
hist_den_rug, ax = plt.subplots() 
ax = sns.distplot(tips['total_bill'], rug=True) ##rug = True로 하여 양탄자 그래프 생성
ax.set_title('Total Bill Histogram with Density and Rug Plot') 
ax.set_xlabel('Total Bill')
```

```python
ax = plt.subplots() 
ax = sns.countplot('day', data=tips) ##이산값을 나타내는 count그래프
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency') 
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) ##seaborn라이브러리로 산점도 그래프
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')  
```

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) ##회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
##산점도 그래프와 히스토그램을 한번에 보여줌 -> jointplot 메서드
```

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
## 육각 그래프 사용
```

```python
kde, ax = plt.subplots() 
ax = sns.kdeplot(data=tips['total_bill'], ##2차원 밀집도 그리기
                 data2=tips['tip'], 
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```python
ax = plt.subplots() 
ax = sns.barplot(x='time', y='total_bill', data=tips) ##barplot 그리기
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

```python
ax = plt.subplots() 
ax = sns.boxplot(x='time', y='total_bill', data=tips) ##boxplot 그리기
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill') 
```

```python
ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', data=tips) ##box에 커널밀도를 추정한 바이올린 그래프
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
fig = sns.pairplot(tips) ##관계 그래프 그리기 -> 중복된 그래프가 표현되는 단점
```

```python
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) 
plt.show()
##중복된 정보를 제거하여 효율적인 그래프 생성
```

```python
violin, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
plt.show() ##바이올린 그래프에 생성추가
```

```python
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False) 
plt.show() ##산점도,관계그래프에 색상 추가,회귀선은 제거
```

```python
sns.set_style('whitegrid')  ##스타일설정
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() ##그래프에 스타일을 적용
```
